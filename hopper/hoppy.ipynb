```python
desired_state = [
    "rosemarie",
    "gracie",
    "shauna",
    "constance",
    "bertram",
    "antoinette",
    "dana",
    "zane",
    "tracie",
    "louie",
    "wes",
    "loretta",
    "chasity",
    "olive",
    "olen",
    "hopper",
]
```

```python
def resetState():
    return {
        'hopper': 0,
        'line': [
            'hopper',
            'wes',
            'gracie',
            'zane',
            'constance',
            'rosemarie',
            'shauna',
            'chasity',
            'louie',
            'tracie',
            'dana',
            'olen',
            'olive',
            'loretta',
            'bertram',
            'antoinette',
        ],
    }
```

```python
def do_hop(state, hoppee):
    hopper = state['hopper']
    line = state['line']

    hops = [
        (hopper - 4, hopper >= 4),
        (hopper - 1, hopper % 4 != 0),
        (hopper + 1, hopper % 4 != 3),
        (hopper + 4, hopper < 12),
    ]

    hoppees = { line[hop]: hop for hop, legal in hops if legal }
    people = ', '.join(hoppees)

    target = hoppees[hoppee]
    line[hopper], line[target] = line[target], line[hopper]
    state['hopper'] = target
    
    print('\t'.join([str(i) for i in range(16)]))
    print('\t'.join([x[:7] for x in desired_state]))
    print('\t'.join([x[:7] for x in state['line']]))

    hopper = state['hopper']
    hops = [
        (hopper - 4, hopper >= 4),
        (hopper - 1, hopper % 4 != 0),
        (hopper + 1, hopper % 4 != 3),
        (hopper + 4, hopper < 12),
    ]

    people = ', '.join({ line[hop]: hop for hop, legal in hops if legal })

    print(f'you can hop with {people}.\n')
    
    return state
```

```python
state = resetState()
state = do_hop(state, "constance")
state = do_hop(state, "rosemarie")
state = do_hop(state, "wes")
state = do_hop(state, "constance")
state = do_hop(state, "rosemarie")
state = do_hop(state, "wes")
state = do_hop(state, "constance")
state = do_hop(state, "gracie")
state = do_hop(state, "gracie")
```
